<h3>Loading data in arrays(Matrix)</h3><br>
Adding bias term (arrays of 1's)<br>
This data is basic data.It's a made up data.So therefore the  accuracy is high

```
 X=np.array([np.ones(data['17.9302012052'].shape),data['17.9302012052'],data['94.5205919533']]).T
 Y=np.array(data['320.259529602']).T
```

Plotting data in 3-D space

```
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(X[:,0],X[:,1],Y)
plt.show() 
```

Calculation of weights<br>
Using normalisation equation

```
#calculate weights

W=np.linalg.solve(np.dot(X.T,X),np.dot(X.T,Y))
#w=np.linalg.pinv(X.T.dot(X)).dot(X.T.dot(Y))

Yhat=np.dot(X,W)
```

Calculation of R-square

```
#r-squared

d1=Y-Yhat
d2=Y-Y.mean()

r2=1-(d1.dot(d1)/d2.dot(d2))
print("Accuracy: ",r2)
```
